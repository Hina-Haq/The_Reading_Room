# 02. Testing for one book

```python
# defining the url of the webpage to be scraped
url = "https://loja.skeelo.com/products/o-mundo-de-sofia"
```

```python
# defining the headers to mimic a browser request
header = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/124.0.0.0 Safari/537.36"
    )
#    , "Accept-Language": "pt-BR,pt;q=0.9",
#    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
#    "Referer": "https://loja.skeelo.com/"
}
```

```python
# getting the page content
response = requests.get(url, headers=header)

# checking if the request was successful
print(response.status_code)

# making the soup object
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
# test for the title
print(f'Title: {soup.find("h1").get_text(strip=True)}')
```

```python
# getting the description
print(f'Description: {soup.find("div", class_="styles_text__Tc8Kj").get_text(separator=" ", strip=True)}')
```

```python
# getting the cover url
img = soup.find("img", class_="styles_productCoverImg__pmqMS")
cover_url = "https://loja.skeelo.com" + img["srcset"].split()[-2]
print(cover_url)
```

```python
# test to see how it the spans look like
spans = soup.find("ul", class_ = "styles_full__Ni0Ch styles_list__8ivnL").find_all("span")

for span in spans:
    print(span.get_text(strip=True))
```

```python
# getting all the elements I need to see if I can find them - all under span, every 2nd element is the label, the next one is the value
spans = soup.find("ul", class_ = "styles_full__Ni0Ch styles_list__8ivnL").find_all("span")

book_data = {}
for i in range(0, len(spans), 2):
    label = spans[i].get_text(strip=True)
    value = spans[i+1].get_text(strip=True)
    book_data[label] = value

print(book_data["Autor (a)"])
print(book_data["Gênero"])
print(book_data["Editora"])
print(book_data["Páginas"])
print(book_data["Ano"])
print(book_data["Idioma"])
print(book_data["ISBN"])
```
