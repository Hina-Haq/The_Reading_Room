# 03. Beautification of Database

```python
books.head()
```

```python
# retting index to be book_id
books.set_index('book_id', inplace=True)
```

```python
# changing num_ratings into INT
books['num_ratings'] = books['num_ratings'].astype('Int64')
```

```python
books.head()
```

```python
# stripping " " from title and author columns
books['title'] = books['title'].str.strip().str.strip('"')
books['author'] = books['author'].str.strip().str.strip('"')
```

```python
# checking if books with " " in the name were stripped correctly
books[books['title'].str.contains('paint', case=False)][['title', 'genres']]
```

```python
# removing duplicated genres from genres_list
def remove_duplicate_genres(genre_list):
    seen = []
    for genre in genre_list:
        if genre not in seen:
            seen.append(genre)
    return seen

def clean_genres_string(genres_string):
    genre_list = genres_string.split(', ')
    unique_genres = remove_duplicate_genres(genre_list)
    return ', '.join(unique_genres)

books['genres_clean'] = books['genres_clean'].apply(clean_genres_string)
```

```python
# checking if duplicated genres were removed correctly
books[books['author'].str.contains('Agatha Christie', case=False)][['title', 'genres_clean']]
```
